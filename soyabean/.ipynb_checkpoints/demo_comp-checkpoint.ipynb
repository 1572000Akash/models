```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
soya1= pd.read_csv("Soyabean1.csv")
```

```python
soya1
```

```python
soya1= soya1.groupby([pd.to_datetime(soya1.Arrival_date).dt.strftime('%b'),'state','variety','district'])['modal_price'].mean().reset_index(name='modal_price')
```

```python
soya1
```

```python
soya1['arrival_date'] = pd.Categorical(soya1['arrival_date'], categories=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],ordered=True)
soya1=soya1.sort_values('arrival_date',ascending=True)
```
