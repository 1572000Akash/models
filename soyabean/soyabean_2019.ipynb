```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sp = pd.read_csv("Rajasthan_top_ten_district.csv")
```

```python
sp.head()
```

```python
x=sp['district'] 
y=sp['esti.prod']
z=sp['model_price']
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = plt.axes(projection='3d')
colors=['red','green','blue','black','yellow','violet','purple','m','orange','white']
```

```python
ax.bar(x,y,z,color=colors,label=colors)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.legend()
plt.show()
```
