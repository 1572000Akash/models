```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('NCDEX_data.csv')
dataset.head(2)
```

```python
#dataset.drop(['Market Name','Commodity', 'Variety','Variety','Grade','Min Price (Rs./Quintal)','Max Price (Rs./Quintal)'], axis = 1,inplace=True)
```

```python
dataset['Date']=pd.to_datetime(dataset['Date'])
```

```python
dataset = dataset.set_index("Date")
```

```python
dataset
```

```python
X = dataset.iloc[:, 0].values
y = dataset.iloc[:,0].values
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X.reshape(-1,1))
y = sc_y.fit_transform(y.reshape(-1,1))
```

## Fit SVR

```python
# Fitting SVR to the dataset
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X, y)
```

```python
# Predicting a new result
#predict data
y_predict = regressor.predict(sc_X.fit_transform(X))
y_predict = sc_y.inverse_transform(y_predict)
```

```python
svr_prediction=pd.DataFrame(y_predict)
```

```python
svr_prediction
```
