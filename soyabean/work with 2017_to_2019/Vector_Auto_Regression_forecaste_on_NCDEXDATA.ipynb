```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read the data
df = pd.read_csv("NCDEX_data.csv")
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
#df['Date'] = pd.to_numeric(df['Date'] , errors = 'coerce')
```

```python
df = df.set_index("Date")
#df['Date']=df.Date.astype(str)
```

```python
df.head()
```

```python
train = df.iloc[0:501].values
valid = df.iloc[500:677].values
```

```python
train
```

```python
valid
```

```python
from statsmodels.tsa.vector_ar.var_model import VAR
model = VAR(train,valid)
model= model.fit()
```

```python
# make prediction on validation
prediction = model.forecast(train, steps=len(valid))
```

```python
pd.DataFrame(prediction)
```
